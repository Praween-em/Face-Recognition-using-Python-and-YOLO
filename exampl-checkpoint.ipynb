```python
from deepface import DeepFace
# Load the image
img_path = "IMG_20240217_134148.jpg"

# Use DeepFace to analyze the image and detect emotions
result = DeepFace.analyze(img_path, actions=['emotion'])

# Print the detected emotions
print("Emotions detected in the image:")
for emotion, score in result['emotion'].items():
    print(f"{emotion}: {score}")
```
